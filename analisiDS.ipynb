# K-Means

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
import seaborn as sns    
from sklearn.decomposition import PCA
import plotly.express as px
from scipy.cluster.hierarchy import dendrogram
from sklearn.cluster import KMeans 
from sklearn import metrics 

df2 = df.drop(df.columns[[0, 1, 2, 3, 4]], axis = 1)
# Inizializzazione dello scaler
scaler = StandardScaler()

# Autoscaling dei dati
df_scaled = scaler.fit_transform(df2)

# Creazione di un nuovo dataframe con i dati autoscalati
df_scaled = pd.DataFrame(df_scaled, columns=df2.columns)

pca = PCA(n_components=2)
pca_df = pca.fit_transform(df_scaled)

pca_dataframe = pd.DataFrame(pca_df, columns = ["PCA1", "PCA2"])
```

```python
kmeans_model = KMeans(n_clusters = 4)
kmeans_model.fit(pca_dataframe)
```

```python
kmeans_model.cluster_centers_.shape
kmeans_model.labels_
pca_dataframe["cluster"] = kmeans_model.labels_
```

```python
fig, ax1 = plt.subplots(1, figsize = (10,8))

sns.scatterplot(x = "PCA1", y = "PCA2", data = pca_dataframe, hue = "cluster", palette = "deep", ax = ax1, legend=False)
plt.title("Clustering ottenuto tramite K-Means")
plt.show()
```

```python
labels_k = kmeans_model.labels_
score_k = metrics.silhouette_score(pca_dataframe.drop("cluster", axis=1), labels_k)
score_k
```

```python
k_to_test = range(2,10,1) # [2,3,4, ..., 24]
silhouette_scores = {}

for k in k_to_test:
    model_kmeans_k = KMeans( n_clusters = k )
    model_kmeans_k.fit(pca_dataframe.drop("cluster", axis = 1))
    labels_k = model_kmeans_k.labels_
    score_k = metrics.silhouette_score(pca_dataframe.drop("cluster", axis=1), labels_k)
    silhouette_scores[k] = score_k
```

```python
plt.figure(figsize = (16,5))
plt.plot(silhouette_scores.values())
plt.xticks(range(0,8,1), silhouette_scores.keys())
plt.title("Silhouette Metric")
plt.xlabel("k")
plt.ylabel("Silhouette")
plt.axvline(1, color = "r")
plt.axvline(2, color = "orange")
plt.show()
```

```python
from sklearn.cross_decomposition import PLSRegression
import numpy as np
from sklearn.model_selection import train_test_split
# Caricamento dei dati
# Assumi che i dati siano caricati in un array numpy chiamato "dataset"
# Assumi che le variabili dipendenti siano collocate nelle colonne 0, 1 e 2
# Assumi che le variabili indipendenti (spettri NIR) siano collocate nelle colonne 3 in poi
df2 = df.drop(df.columns[[0, 1]], axis = 1)
# Inizializzazione dello scaler
scaler = StandardScaler()

# Autoscaling dei dati
df_scaled = scaler.fit_transform(df2)

# Creazione di un nuovo dataframe con i dati autoscalati
df_scaled = pd.DataFrame(df_scaled, columns=df2.columns)

# Estrazione delle variabili indipendenti (spettri NIR)
X = df_scaled.iloc[:, 3:-1]

# Estrazione delle variabili dipendenti (NOH, RS e NAC)
Y = df_scaled.iloc[:,0:3]

# Suddivisione del dataset in dati di addestramento e di test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Creazione del dataset di test combinando le variabili dipendenti e indipendenti
test_data = np.concatenate((Y_test, X_test), axis=1)

# Inizializzazione del modello PLS con 7 componenti latenti
pls = PLSRegression(n_components=7)

# Addestramento del modello PLS
pls.fit(X_train, Y_train)

# Previsioni su nuovi dati
#  i dati di test siano caricati in un array numpy chiamato "test_data"
X_test = test_data[:, 3:]
predictions = pls.predict(X_test)

# Valutazione del modello
# le variabili dipendenti di test siano collocate nelle colonne 0, 1 e 2
Y_test = test_data[:, 0:3]

# Calcolo dell'errore RMSE
rmse = np.sqrt(np.mean((predictions - Y_test) ** 2))
print("RMSE: ", rmse)
```

```python

def column(matrix, i):
    return [row[i] for row in matrix]
```

```python
import matplotlib.pyplot as plt

# Supponendo che il modello PLS sia stato addestrato e le previsioni siano state calcolate come "predictions"
# Supponendo che i valori effettivi per la variabile dipendente NOH siano stati estratti come "Y_test_NOH"

# Creazione di un array con gli indici dei campioni
indices = range(len(predictions))

# Plot dei valori effettivi e predetti per NOH
plt.scatter(indices, column(Y_test, 0), label='Effettivi')
plt.scatter(indices, column(predictions, 0), label='Predetti')
plt.plot(indices, column(Y_test, 0), label='Effettivi')
plt.plot(indices, column(predictions, 0), label='Predetti')

# Personalizzazione del plot
plt.xlabel('Campioni')
plt.ylabel('NOH')
plt.title('Confronto tra valori effettivi e predetti per NOH')
plt.legend()

# Mostra il plot
plt.show()
```
