```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

df = pd.read_csv('resine.csv')
plt.figure(figsize=(10, 5))
df['NOH'].plot(kind='hist', rwidth=0.9)
plt.ylabel("Frequenza", fontsize=20)
plt.xlabel("NOH", fontsize=20)
plt.title("Frequenze NOH", fontsize=20)
```

```python
plt.figure(figsize=(10, 5))
df['NAC'].plot(kind='hist', rwidth=0.9)
plt.ylabel("Frequenza", fontsize=20)
plt.xlabel("NAC", fontsize=20)
plt.title("Frequenze NAC", fontsize=20)
```

```python
plt.figure(figsize=(10, 5))
df['RS'].plot(kind='hist', rwidth=0.9)
plt.ylabel("Frequenza", fontsize=20)
plt.xlabel("RS", fontsize=20)
plt.title("Frequenze RS", fontsize=20)
```

# Boxplot

```python
plt.figure(figsize=(10, 5))
boxplot = df.boxplot(column=["NOH", "RS", "NAC"])
plt.title("Analisi dei boxplot", fontsize=20)
```

# Matrix plot

```python
plt.figure(figsize=(20, 15))
fig = px.scatter_matrix(df,
    dimensions=["NOH", "RS", "NAC",], color="FAM")
fig.show()
```
