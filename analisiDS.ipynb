```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

df = pd.read_csv('resine.csv')
plt.figure(figsize=(10, 5))
df['NOH'].plot(kind='hist', rwidth=0.9)
plt.ylabel("Frequenza", fontsize=20)
plt.xlabel("NOH", fontsize=20)
plt.title("Frequenze NOH", fontsize=20)
```

```python
plt.figure(figsize=(10, 5))
df['NAC'].plot(kind='hist', rwidth=0.9)
plt.ylabel("Frequenza", fontsize=20)
plt.xlabel("NAC", fontsize=20)
plt.title("Frequenze NAC", fontsize=20)
```

```python
plt.figure(figsize=(10, 5))
df['RS'].plot(kind='hist', rwidth=0.9)
plt.ylabel("Frequenza", fontsize=20)
plt.xlabel("RS", fontsize=20)
plt.title("Frequenze RS", fontsize=20)
```

# Boxplot

```python
plt.figure(figsize=(10, 5))
boxplot = df.boxplot(column=["NOH", "RS", "NAC"])
plt.title("Analisi dei boxplot", fontsize=20)
```

# Matrix plot

```python
plt.figure(figsize=(20, 15))
fig = px.scatter_matrix(df,
    dimensions=["NOH", "RS", "NAC",], color="FAM")
fig.show()
```

```python
# filtro il dataset mantenendo solo le colonne dello spettro NIR
plt.figure(figsize=(10, 5))
df_nir = df.drop(df.columns[[0, 1, 2, 3, 4]], axis = 1)
df_nir.iloc[0].plot()
plt.ylabel("Values", fontsize=20)
plt.xlabel("Lunghezze d'onda", fontsize=20)
plt.title("Plot del primo spettro NIR", fontsize=20)
```

```python
import matplotlib.pyplot as plt
import numpy as np
num_plots=99
df_nir_noh = df.sort_values('NOH').iloc[0:22]
df_nir_noh = df_nir_noh.drop(df_nir_noh.columns[[0, 1, 2, 3, 4]], axis = 1)

plt.figure(figsize=(30, 15))


for index, row in df_nir_noh.iterrows():
    plt.plot(row)

plt.ylabel("Values", fontsize=20)
plt.xlabel("Lunghezze d'onda", fontsize=20)
plt.title("Plot degli spettri NIR totali", fontsize=20)
plt.show()
```

```python
df2 = df[df["FAM"]==1]
columns = ["NOH"] + [f"V{i}" for i in range(1, 100)]
lines = df2[columns].sort_values("NOH")

# Crea una mappa di colori
cmap = plt.get_cmap('RdPu')

# Ottieni i valori della colonna "NOH"
noh_values = lines["NOH"].values

columns_to_plot = lines.columns[lines.columns != "NOH"]
plt.figure(figsize=(25, 10))
# Itera sulle righe del dataframe
for index, row in lines.iterrows():
    # Ottieni i valori delle colonne selezionate nella riga corrente
    values = row[columns_to_plot].to_list()
    
    # Calcola l'indice di colore in base al valore di "NOH"
    color_index = np.interp(row["NOH"], (noh_values.min(), noh_values.max()), (0, 1))
    color_rgb = cmap(color_index)
    # Plotta i valori colorando la linea
    plt.plot(values, color = color_rgb, label=f'{"NOH:" + str(row["NOH"])}')

# Aggiungi legenda e titolo al grafico
plt.legend(ncol=len(df2.index)/3, prop={'size': 15})
plt.title('Plot degli spettri NIR della prima famiglia (colore in base a NOH)', fontsize=20)
plt.ylabel("Values", fontsize=20)
plt.xlabel("Lunghezze d'onda", fontsize=20)
# Mostra il grafico
plt.show()
```
